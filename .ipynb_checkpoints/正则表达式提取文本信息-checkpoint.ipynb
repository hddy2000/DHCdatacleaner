```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
from scipy.sparse import issparse
import seaborn as sns
```

```python
path='D:/Work/Data/RawData/analysis/caichao/'#input数据的path
data_ori = pd.read_csv(path+'caichao1.0.csv',encoding='gbk',index_col=0)
```

```python
data_ori.index = range(len(data_ori))
df = data_ori[0:400]
```

```python
import re

#长度为2，且四舍五入（Rounding）
#长度不为2，val置零

def func_Round(x):
    if (len(x) == 2 and int(x[1]) > 3):
        val = 1
    else:val = 0
    return val

def Rounding(col,vals,method = func_Round):
    #新设周次和天次的临时列
    df[col+'temp1'] = pd.Series()
    df[col+'temp2'] = 0
    #提取的信息‘四舍五入’
    df[col+'temp1'] = vals.map(lambda x: list(x)[0])
    df[col+'temp2'] = vals.map(lambda x: method(x))
    df[col+'temp1'] = df[col+'temp1'].astype('float64')
    df[col+'temp2'] = df[col+'temp2'].astype('float64')
    df['Round' + col] = df[col+'temp1'] + df[col+'temp2']
    return df['Round' + col]

def Text_extraction(df,cols,re_method,Round=False,method = func_Round):
    for col in cols:

        #非空行的索引
        index = df[col].isnull().values==False 
        try:
            #非空的文字信息提取
            val_re = df[col][index].map(lambda x: re.findall(re_method,x))
            df['re_'+col] = val_re
        except:
            #这里可以调用下已经写好的函数，将这一列中有问题的值暴露出来
            print (col,'提取信息异常')
        if Round:
            df['Round' + col] = Rounding(col, val_re, method = func_Round)
    return df

```

```python
cols = [u'计孕周']
Text_extraction(df,cols,re_method = r"\d+",Round=True,method = func_Round)
```

```python
cols = [u'胎盘直径',u'胎盘厚度']
Text_extraction(df,cols,re_method = r"\d+")
```

```python
#删除异常值,方法是取区间内的值
def del_value(input_dataframe,sel_cols,minv,maxv,copy=False):
    '''
    :param input_dataframe: pandas.DataFrame
    :param sel_cols: list, selected columns need to replace,
    :param minv: Lower boundary value，
    :param maxv: Upper boundary value,
    :param copy:  copy the data or not,
    :return: input_dataframe
    '''
    if copy:
        input_dataframe=input_dataframe.copy()
    index = (input_dataframe[sel_cols]>minv) & (input_dataframe[sel_cols]<maxv)
    output_dataframe = input_dataframe[index]
    return output_dataframe
```
